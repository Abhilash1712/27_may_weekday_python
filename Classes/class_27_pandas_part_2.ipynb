```python
space_org = {'Organizations_name' : ["CNES", "ISRO", "NASA", "JAXA", "CNSA"],
                   'Headquarters' : ["Paris", "Bangalore", "Washington",
                                     "Tokyo", "Beijing"],
                   'Country_name' : ["France", "India", "USA", "Japan", "China"],
                 'Formation_year' : [1961, 1969, 1958, 2003, 1993]
             }
```

```python
df = pd.DataFrame(space_org)
```

```python
df
```

```python
for i in df.itertuples():
    print(i,end='\n\n')
```

```python
for i in df.itertuples():
    print(i.Organizations_name,end='\n\n')
```

```python
for i in df.itertuples():
    if i.Country_name=="France":
        print(i.Headquarters,end="\n\n")
```

```python
for i in df.iterrows():  #(index, series)
    print(i[1]['Headquarters'],end='\n\n')
```

```python
for i in df.iterrows():  #(index, series)
    print(i,end='\n\n')
```

```python
for i,j in df.iterrows():  #(index, series)
    print(j['Headquarters'],end='\n\n')
```

```python
for i in df.iteritems():  #(column_name, data)
    print(i,end='\n\n')
```

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
pd.concat([india_weather, us_weather], ignore_index=True)
```

```python
pd.concat([india_weather, us_weather], axis=1)
```

```python
# 2 level indexing

df = pd.concat([india_weather, us_weather], keys = ['india','us'])
```

```python
df.loc['us',1]
```

# Merging

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
pd.merge(df1, df2, how='inner')
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how='inner')
```

```python
pd.merge(df1, df2, how='left')
```
