```python
# Iteration
```

```python
array = np.array([4,5,6,7,3,2,4,8,6,5,4,3]).reshape(4,3)
array
```

```python
array = np.array([4,5,6,7])
```

```python
for i in array:
    print(i)
```

```python
array = np.array([4,5,6,7,3,2,4,8,6,5,4,3]).reshape(4,3)
for i in array:
    print(i)
```

```python
for i in array:
    for j in i:
        print(j)
```

```python
for i in array.flatten():
    print(i)
```
