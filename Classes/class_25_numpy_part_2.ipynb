```python
# copy and view
```

```python
a = np.array([1,2,3,4,5])
x = a.copy()
x[0] = 42

print(a)
print(x)
```

```python
a = np.array([1,2,3,4,5])
x = a.view()
x[0] = 42

print(a)
print(x)
```

```python
a = np.array([1,2,3,4,5])
x = a
x[0] = 42

print(a)
print(x)
```

```python
the main difference between copy and view of an arrya is that te copy is a new array whereas a view is just
a view of the original array
```
