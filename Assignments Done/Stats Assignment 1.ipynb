```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

## Line Graph

```python
x = np.arange(1,11)
y = x**2
```

```python
plt.title("Number vs it's Square")
plt.xlabel('Numbes')
plt.ylabel('Square of the Numbers')

plt.plot(x,y,'rv:')
```

##  Piechart

```python
Expenditute = np.array([45,10,15,25,30])
Exp_labels = ['Car price','Road Tax','Accessories','Tyres','Modes Upgrade']
```

```python
plt.title('Total Car Expenditure')
plt.pie(Expenditute,labels = Exp_labels , autopct='%0.2f%%')
```

### Donut Chart

```python
Expenditute = np.array([45,10,15,25,30])
Exp_labels = ['Car price','Road Tax','Accessories','Tyres','Modes Upgrade']
```

```python
plt.title('Total Car Expenditure')
plt.pie(Expenditute,labels = Exp_labels , autopct='%0.2f%%')
c_circle = plt.Circle((0,0),0.6,color='w')
plt.gcf()
plt.gca().add_artist(c_circle)
```

### Bar Graph

```python
Company = ['Jeep','Ford','Tesla','Toyota','BMW','Daimler']

Sales_per_month = np.array([10000,25000,7000,40000,15000,18000])

plt.title('Sales in the Month of July 2022')
plt.bar(Company,Sales_per_month)
```

### Scatter Plot

```python
x = np.array([10,20,30,40,60,30,33,55,66,78,34,78,90,43,67,12])
y = np.array([500,700,800,2000,3800,900,1000,1800,2500,3700,2400,2300,5700,6900,3300,300])

plt.ylabel('Income')
plt.xlabel('Age')
plt.scatter(x,y)
```

### Histogram

```python
blood_sugar_men = np.array([113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129])
blood_sugar_women = np.array([67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100])
```

```python
plt.hist([blood_sugar_men,blood_sugar_women], bins=[70,100,125,150],color=['b','r'], label=['Men','Women'])
plt.legend()
plt.title('Blood Sugar Level')
```
