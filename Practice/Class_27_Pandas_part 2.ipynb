## Iterate over Dataframes

```python
space_org = {'Organizations_Names' : ["CNES","ISRO","NASA","JNXA","CNSA"],
             'Headquater'          : ["Paris", "Banglore", "Washington","Tokyo","Beijing"],
             'Country_name'        : ["France","India","USA","JApan","China"],
             'Formation_Year'      : [1961,1969,1958,2003,1993]}
```

```python
df = pd.DataFrame(space_org)
df
```

```python
# itertuples

for i in df.itertuples():
    print(i,end='\n\n')
```

```python
for i in df.itertuples():
    print(i.Organizations_Names,end='\n\n')
```

```python
for i in df.itertuples():
    if i.Country_name=='France':
        print(i.Headquater,end='\n\n')
```

```python
# iterrows

for i in df.iterrows():
    print(i,end='\n\n')
```

```python
for i in df.iterrows():
    print(i[1]['Headquater'] ,end='\n\n')
```

```python
# iteritems

for i in df.iteritems():
    print(i,end='\n\n')
```

```python
for i in df.iteritems():
    print(i[0],end='\n\n')
```

## Concatination

```python
india_weather = pd.DataFrame({
               "City" : ["Mumbai", "Delhi" , "Banglore"],
               "Temperature" : [32,45,30],
               "HUmidity" : [80,60,78],
})
india_weather
```

```python
us_weather = pd.DataFrame({
               "City" : ["New York", "Chicago" , "Orlando"],
               "Temperature" : [21,14,35],
               "HUmidity" : [68,65,75],
})
us_weather
```

```python
pd.concat([india_weather,us_weather])
```

```python
pd.concat([india_weather,us_weather], axis =1)    # Column wise
```

```python
pd.concat([india_weather,us_weather], ignore_index= True)
```

```python
# 2 level indexing


df = pd.concat([india_weather,us_weather], keys = ['India','US'])
df
```

```python
df.loc['US'].loc[1]
```

```python
df.loc['US',1]
```

```python
temperature_df = pd.DataFrame({
               "City" : ["Mumbai", "Delhi" , "Banglore"],
               "Temperature" : [32,45,30],
               
}, index=[0,1,2])
temperature_df
```

```python
pd.concat([temperature_df,Windspeed_df])
```

```python
pd.concat([temperature_df,Windspeed_df], axis=1)
```

## Merging

```python
df1 = pd.DataFrame({
               "City" : ["New York", "Chicago" , "Orlando"],
               "Temperature" : [21,14,35],
              
})
df1
```

```python
df2 = pd.DataFrame({
               "City" : ["Chicago", "New York" , "Orlando"],
        
               "HUmidity" : [65,68,75],
})
df2
```

```python
pd.merge( df1 , df2 , how = 'inner')
```

```python
df1 = pd.DataFrame({
               "City" : ["New York", "Chicago" , "Orlando","Baltimore"],
               "Temperature" : [21,14,35,38],
              
})
df1
```

```python
df2 = pd.DataFrame({
               "City" : ["Chicago", "New York" ,"Sandiego"],
        
               "HUmidity" : [65,68,71],
})
df2
```

```python
pd.merge( df1 , df2 , how = 'inner')
```

```python
pd.merge( df1 , df2 , how = 'left')
```

```python
pd.merge( df1 , df2 , how = 'right')
```

```python
pd.merge( df1 , df2 , how = 'outer')
```

```python
pd.merge( df1 , df2 , how = 'outer', indicator=True)
```
