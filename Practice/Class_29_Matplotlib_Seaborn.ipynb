## Piechart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(exp_vals, labels = exp_labels)
```

```python
plt.pie(exp_vals, labels = exp_labels , autopct='%0.2f%%')
```

```python
plt.pie(exp_vals, labels = exp_labels , autopct='%0.2f%%',radius=1.5)
```

```python
plt.pie(exp_vals, labels = exp_labels , autopct='%0.2f%%',radius=1.5, explode=[0.04,0,0,0,0.08])
```

# Seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# Relational Plot (relplot)
```

```python
sns.relplot( x = 'total_bill', y = 'tip' , data = tips)
```

```python
sns.relplot( x = 'total_bill', y = 'tip' , data = tips, hue ='smoker')   # hue --> categorical --> based on colour
```

```python
sns.relplot( x = 'total_bill', y = 'tip' , data = tips, hue ='smoker' ,style='time')    # style ---> categorical --> based on style     
```

```python
sns.relplot( x = 'total_bill', y = 'tip' , data = tips, hue ='smoker' ,style='time',size='size')
```

```python
sns.relplot( x = 'total_bill', y = 'tip' , data = tips, hue ='size' ,style='time',size='smoker')
```

```python
sns.relplot( x = 'total_bill', y = 'tip' , data = tips, hue ='size' ,style='time',size='smoker',col='sex')
```

```python
# categorical Plot
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
sns.catplot(x='total_bill',y='day',data=tips,kind='box')
```

```python
# Distribution Plot
```

```python
sns.displot(x=tips['total_bill'])
```

```python
sns.displot(x=tips['total_bill'], kde = True)   # kernel density estimator
```

```python
sns.distplot(x=tips['total_bill'], kde = True, hist = False)
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'])
```
