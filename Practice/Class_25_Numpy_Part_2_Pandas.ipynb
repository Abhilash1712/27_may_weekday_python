```python
# Copy and View
```

```python
a = np.array([1,2,3,4,5])
x = a.copy()                 # creating a separate element 
x[0] = 42

print(a)
print(x)
```

```python
a = np.array([1,2,3,4,5])
x = a.view()               # kind of aliasing
x[0] = 42

print(a)
print(x)
```

```python
a = np.array([1,2,3,4,5])
x = a
x[0] = 42

print(a)
print(x)
```

```python
# Using Dictionary

data = {'name1':'afsan','name2':'learnbay','name3':'google'}
pd.Series(data)
```

```python
data = {'name1':'afsan','name2':'learnbay','name3':'google'}
df = pd.Series(data)
```

```python
data = {'name1':'afsan','name2':'learnbay','name3':'google'}
df = pd.Series(data, index = [0, 'name1', 2])
df
```

```python
# Indexing

# 1. zero indexing
# 2. custom indexing


df[1]  # zero indexing
```

```python
df['name3']   # custom indexing
```

```python
# Slicing


data = {'name1':'afsan','name2':'learnbay','name3':'google'}
df = pd.Series(data)
df
```

```python
df['name1':'name2']     # custom indexing
```
