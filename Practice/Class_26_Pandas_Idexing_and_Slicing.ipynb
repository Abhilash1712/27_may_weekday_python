## Slicing

```python
# 1. Normal Way ( do slicing only on rows )

df[100:200]    # Zero indexing
```

```python
df[100:200:2]
```

```python
index = ['row1','row2','row3','row4','row5','row6']
new_data = df[0:6]
new_data
```

```python
new_data.index = index
new_data
```

```python
new_data[0:3]       # Zero Indexing
```

```python
new_data['row1':'row4']     # Custom Indexing
```

```python
new_data
```

```python
# loc   (location)

# 1. can do slicing both on rows and column
# 2. you can only do custom indexing

# syntax( start row:stop row , start column : stop cloumn)
```

```python
new_data.loc['row1':'row4']
```

```python
new_data = df[0:6]
new_data
```

```python
new_data.loc[0:3]     # it is also an custom indexing row 3 is included in this kind of slicing
```

```python
new_data.index = index
new_data
```

```python
new_data.loc[::2 ,::2]
```

```python
new_data
```

```python
new_data.loc[['row1','row3','row4'],['tripduration','stoptime','bikeid']]
```
